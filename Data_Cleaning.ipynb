#changing codes to labels

```python
ds['Accident_Severity'].replace([1, 2, 3],['Fatal', 'Serious', 'Slight'], inplace=True)
ds['Day_of_Week'].replace([1,2,3,4,5,6,7],['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'], 
                          inplace=True)
ds['Urban_or_Rural_Area'].replace([1, 2], ['Urban', 'Rural'], inplace=True)
ds['Road_Type'].replace([1, 2, 3, 6, 7, 9], ['Roundabout', 'One way street', 'Dual carriageway', 'Single carriageway',
                                            'Slip road', 'Unknown'], inplace=True)
ds['Junction_Detail'].replace([3,  2,  6,  0,  8,  1,  7,  9,  5, -1], ['T or staggered junction', 'Mini-roundabout', 'Crossroads',
                                                                       'Not at junction or within 20 metres', 'Private drive or entrance',
                                                                       'Roundabout', 'More than 4 arms (not roundabout)', 'Other junction',
                                                                       'Slip road', 'Data missing or out of range'], inplace=True)
ds['Junction_Control'].replace([4,  2, -1,  3,  1], ['Give way or uncontrolled', 'Auto traffic signal', 'Data missing or out of range',
                                                    'Stop sign', 'Authorised person'], inplace=True)
ds['Pedestrian_Crossing-Human_Control'].replace([0,  1,  2, -1], ['None within 50 metres', 'Control by school crossing patrol',
                                                                 'Control by other authorised person', 'Data missing or out of range'], 
                                                inplace=True)
ds['Pedestrian_Crossing-Physical_Facilities'].replace([0,  1,  5,  4,  8,  7, -1], ['No physical crossing facilities within 50 metres',
                                                                                    'Zebra', 'Pedestrian phase at traffic signal junction',
                                                                                   'Pelican, puffin, toucan or similar non-junction pedestrian light crossing',
                                                                                   'Central refuge', 'Footbridge or subway', 'Data missing or out of range'], 
                                                      inplace=True)
ds['Did_Police_Officer_Attend_Scene_of_Accident'].replace([1,  2,  3, -1], ['Yes', 'No', 'accident was reported using a self completion form',
                                                                           'data missing'], inplace=True)
ds['Light_Conditions'].replace([4, 1, 6, 7, 5], ['Darkness - lights lit', 'Daylight', 'Darkness - no lighting',
                                                'Darkness - lighting unknown', 'Darkness - lights unlit'], inplace=True)
ds['Weather_Conditions'].replace([1, 2, 8, 7, 5, 9, 4, 6, 3], ['Fine no high winds', 'Raining no high winds', 'Other',
                                                              'Fog or mist', 'Raining + high winds', 'unknown',
                                                               'Fine + high winds', 'Snowing + high winds', 
                                                               'Snowing no high winds'], inplace=True)
ds['Road_Surface_Conditions'].replace([1,  2,  4,  3,  5, -1], ['Dry', 'Wet or damp', 'Frost or ice', 'Snow', 
                                                                'Flood over 3cm deep', 'Data missing or out of range'], 
                                      inplace=True)
ds['Special_Conditions_at_Site'].replace([ 0,  4,  2,  1,  6,  5,  3,  7, -1], ['None', 'Roadworks', 'Auto signal part defective',
                                                                               'Auto traffic signal out', 'Oil or diesel',
                                                                               'Road surface defective', 'Road sign or marking defective or obscured',
                                                                               'Mud', 'Data missing or out of range'], inplace=True)
ds['Carriageway_Hazards'].replace([0,  2,  6,  7,  1,  3, -1], ['None', 'Other object on road', 'Pedestrian in carriageway',
                                                               'Any animal in carriageway', 'Vehicle load on road',
                                                               'Previous accident', 'Data missing or out of range'], 
                                  inplace=True)
```

```python
police_force= pd.read_csv(r'C:\Users\mycom\Desktop\Police_Force.csv')
ds['Police_Force']  = ds['Police_Force'].map(police_force.set_index('code').to_dict()['label'])
```

```python
district= pd.read_csv(r'C:\Users\mycom\Desktop\Local_Authority_District.csv')
ds['Local_Authority_(District)'] = ds['Local_Authority_(District)'].map(district.set_index('code').to_dict()['label'])
```

```python
highway= pd.read_csv(r'C:\Users\mycom\Desktop\Local_Authority_Highway.csv')
ds['Local_Authority_(Highway)']  = ds['Local_Authority_(Highway)'].map(highway.set_index('Code').to_dict()['Label'])
```
