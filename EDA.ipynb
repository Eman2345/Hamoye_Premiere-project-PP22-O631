## Number of Casualties by Day of Week

```python
uk_accident[['Number_of_Casualties', 'Day_of_Week']].groupby('Day_of_Week').sum()
```

## Number of Casualties by Month
september which is the month of the highest no. of accidents (12396) has (16294) casualties, while july has the highest no. of casualties(16659).

```python
pd.to_datetime(uk_accident.Date)
uk_accident['Date'] = pd.to_datetime(uk_accident['Date'])
uk_accident['month'] = uk_accident['Date'].dt.month
```

```python
sns.barplot(x='month',y='Number_of_Casualties', data=uk_accident, estimator=np.sum)
plt.ylabel("Number_of_Casualties")
plt.xlabel("Month")
plt.show()
```

```python
uk_accident[['Number_of_Casualties', 'month']].groupby('month').sum()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=uk_accident,x='month', y='Number_of_Casualties', marker='o', estimator=np.sum)
```

from february to july, there was a continuous increase in the number of casualties!
from september to december, there was a continuous decrease in the number of casualties!

# Why?

```python
sns.catplot(data=uk_accident, x='month', hue='Junction_Control', y='Number_of_Casualties', estimator=np.sum, kind='point')
```

the number of casualties in those months icrease and decrese proportionally to the no. of casualties occured due to give way or uncontrolled junction.

```python
plt.figure(figsize=(12,6))
sns.catplot(data=uk_accident, x='month', hue='Police_Force', y='Number_of_Casualties', estimator=np.sum, kind='point')
```

the number of casualties in those months icrease and decrese proportionally to the no. of casualties occured at the area controlled by the metropolitan police (Greater London area, (excluding City of London)).

```python
plt.figure(figsize=(12,6))
sns.catplot(data=uk_accident, x='month', hue='Road_Type', y='Number_of_Casualties', estimator=np.sum, kind='point')
```

The number of casualties in those months increase and decrese proportionally to the no. of casualties occured at single carriageway.

```python
plt.figure(figsize=(12,6))
sns.catplot(data=uk_accident, x='month', hue='Junction_Detail', y='Number_of_Casualties', estimator=np.sum, kind='point')
```

The number of casualties in those months increase and decrese proportionally to the no. of casualties occured at T or staggered Junction and roundabout junction.

```python
plt.figure(figsize=(12,6))
sns.catplot(data=uk_accident, x='month', hue='Pedestrian_Crossing-Physical_Facilities', y='Number_of_Casualties', estimator=np.sum, kind='point')
```

The number of casualties in those months increase and decrese proportionally to the no. of casualties occured due to no physical crossing facilities.

```python
plt.figure(figsize=(12,6))
sns.catplot(data=uk_accident, x='month', hue='Pedestrian_Crossing-Human_Control', y='Number_of_Casualties', estimator=np.sum, kind='point')
```

The number of casualties in those months increase and decrese proportionally to the no. of casualties occured due to no one is responsible for pedestrian human crossing.

```python
plt.figure(figsize=(12,6))
sns.catplot(data=uk_accident, x='month', hue='Weather_Conditions', y='Number_of_Casualties', estimator=np.sum, kind='point')
```

```python
plt.figure(figsize=(12,6))
sns.catplot(data=uk_accident, x='month', hue='Road_Surface_Conditions', y='Number_of_Casualties', estimator=np.sum, kind='point')
```

From september to december,
Although the no. of casualties due to (wet/damp) road surface, and (raining no high winds) weather has increased, the total no. of accidents happened in those months decreased.
